#projection of vector

Projection of a vector onto v vector
Formula: Proja = (a.v/||v||^2)*v

```python
#Projection of a on v
a = np.array([2,5])

v = np.array([8,-6])

# firstly we have to find the mangnitude of v vector
magnitude_of_v = np.sqrt(np.sum(v**2))

proj_of_a_on_v = (np.dot(a,v)/magnitude_of_v**2)*v
print("projection of a vector on v vector = ",proj_of_a_on_v)
```

```python
#Projection of a on v
a = np.array([23,45,62])

v = np.array([45,82,67])

# firstly we have to find the mangnitude of v vector
magnitude_of_v = np.sqrt(np.sum(v**2))

proj_of_a_on_v = (np.dot(a,v)/magnitude_of_v**2)*v
print("projection of a vector on v vector = ",proj_of_a_on_v)
```

```python
a = np.array([[2,3],[4,5]])
b = np.array([[6,7],[8,9]])
print(a)
print(b)
sum = a+b
print(sum)
```

**Matrix Subtraction:** Two matrix can be subtracted, only they have the same shape, that is, both the matrix have the same number of rows and columns

```python
a = np.array([[2,3],[4,5]])
b = np.array([[6,7],[8,9]])
```

```python
diff = a-b
print(diff)
```

```python
difference = np.subtract(a,b)
print(difference)
```
