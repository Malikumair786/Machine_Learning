Workflow:

Dataset --> Data preprocessing --> train test split --> logistic regression --> New data --> Benign or maligant

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
cancer_dataset = pd.read_csv('/content/data.csv')
```

```python
print(cancer_dataset)
```

```python
cancer_dataset.head()
```

```python
cancer_dataset = cancer_dataset.drop(['Unnamed: 32','id'], axis = 1)
```

```python
cancer_dataset.head()
```

```python
cancer_dataset.describe()
```

```python
cancer_dataset['diagnosis'].value_counts()
```

```python
cancer_dataset.replace({'diagnosis' :{'B': 1, 'M': 0}},inplace = True)
```

```python
cancer_dataset.head()
```

```python
cancer_dataset['diagnosis'].value_counts()
```

1 --> Benign

0 --> Malingant

```python
cancer_dataset.groupby('diagnosis').mean()
```

Seperating the featrures and target

```python
X = cancer_dataset.drop(columns = 'diagnosis', axis = 1)
Y = cancer_dataset['diagnosis']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into trian and test data

```python
X_train,X_test,y_train,Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training:

Logistic regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy on trianing data 
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(y_train,X_train_prediction)
```

```python
print("The accyracy on training data is : ",training_data_accuracy)
```

```python
#accuracy on test data 
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test,X_test_prediction)
print("The accyracy on test data is : ",test_data_accuracy)
```
